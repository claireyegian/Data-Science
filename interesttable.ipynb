## Interest Table
This notebook calculates the total value of an investment that started at 1000 dollars and is compounded annually using the formula: 
$Interest =\ 1000(1.035^{t})$
where t is years.

```python
from numpy import radians, sin
from numpy import linspace
import matplotlib.pyplot as plt
import pandas as pd
```

```python
interest = [(x,1000*(1.035**x),"$"+str(round(1000*(1.035**x)))) for x in linspace(0,30,31)]
print(interest)
```

```python
df = pd.DataFrame(interest)
df.columns = ['years','value_float','value_str']
```

```python
df
```

```python
plt.plot(df.years, df.value_float)
plt.xlabel(df.years.name)
plt.ylabel('total')
plt.legend('$')
plt.show()
```
